```python
from numpy.random import normal
import scipy.stats as stats
import random
import warnings
from scipy.stats import anderson_ksamp
from statsmodels import robust
from scipy.stats import multivariate_normal
```

```python
def power_l(a, b, g, size=1):
    """Power-law gen for pdf(x)\propto x^{g-1} for a<=x<=b"""
    r = np.random.random(size=size)
    ag, bg = a**g, b**g
    return (ag + (bg - ag)*r)**(1./g)
```

```python
def dist_gen(data, dist):
    warnings.filterwarnings(action = 'once')
    
    mean = np.mean(data)
    sd = np.std(data)
    
    a = (mean / sd)**2
    b = (sd**2) / mean
    
    r = []
    for i in range(1,1000):
        
        if dist == "norm":
            dat = random.sample(list(data), 200)
            gen = np.random.normal(mean, sd, size = 200)
        elif dist == "gam":
            dat = random.sample(list(data), 200)
            gen = np.random.gamma(a, b, size = 200)
        elif dist == "power":
            dat = random.sample(list(data), 200)
            gen = power_l(a=-2, b=np.min(list(data)), g=np.max(list(data))-np.min(list(data)),size= 200)

        s = anderson_ksamp([dat, gen])
        s = s[2]
        r.append(s)

    return(np.mean(r))    
```

```python
def multi_probs(data_Q):
    var = multivariate_normal(mean=[np.mean(data_Q.page), np.mean(data_Q.iMean)], cov=np.cov(data_Q.iMean, data_Q.page))
    probs = []
    for i, j in zip(list(data_Q.page), list(data_Q.iMean)):
        prob = var.pdf([[i,j]])
        probs.append(prob)
    return probs
```

```python
def get_top(prec, data, var):
    zero = list(data.small_path[(data.num_events == 1) & (data[var] < np.percentile(data[var] , prec))])
    first = list(data.small_path[(data.num_events == 2) & (data[var] < np.percentile(data[var] , prec))])
    second = list(data.small_path[(data.num_events == 3) & (data[var] < np.percentile(data[var] , prec))])
    third = list(data.small_path[(data.num_events == 4) & (data[var] < np.percentile(data[var] , prec))])
    total = list(data.small_path)
    remove_ls = first+second+third+zero
    
    ls = [tot for tot in total if not any(remove in tot for remove in remove_ls)]
    data_2 = data[data["small_path"].isin(ls)]
    
    return data_2
```

```python
def models(n,r1,r2,r3):
    y1 = n+n*(n-r1)+(n*(n-r1)-r2)+((n*(n-r1)-r2)-r3)
    y2 = n+n**2+n**3+(n-r1)**4
    y3 = (n-r1)+(n-r1)**2+(n-r1)**3+(n-r1)**4
    y4 = n+n**2+n**3+n**4
    return(y1,y2,y3,y4)
    
```
